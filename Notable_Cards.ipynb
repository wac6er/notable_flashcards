```python
chapter = '7'
textbook = 'Campbell Biology'
number_of_terms = 13

import pandas as pd
from bs4 import BeautifulSoup
from googlesearch import search 
import requests
import re

# to search 
query = 'test+1+uva+comm+law+quizlet'

links = []
for j in search(query, tld="co.in", num=50, stop=50, pause=2): 
    links.append(j)

headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:80.0) Gecko/20100101 Firefox/80.0'}

sub_dict = {'question': [], 'answer': []}

for url in links:
    soup = BeautifulSoup(requests.get(url, headers=headers).content, 'html.parser')


    for i, (question, answer) in enumerate(zip(soup.select('a.SetPageTerm-wordText'), soup.select('a.SetPageTerm-definitionText')), 1):
        sub_dict['question'].append(question.get_text(strip=True, separator='\n'))
        sub_dict['answer'].append(answer.get_text(strip=True, separator='\n'))

df = pd.DataFrame(sub_dict)
df = df.apply(lambda x: x.astype(str).str.lower())
df_final = df.groupby(['question'])['answer'].count().reset_index(name='count')
df_ff = df.merge(df_final, on = 'question', how = 'inner').sort_values(['count'], ascending=False)

df_ff = df_ff.drop_duplicates(subset='question', keep='first', inplace=False)
df_num = df_ff.head(number_of_terms)
df_num.head(number_of_terms)
```

```python
chapter = '3'
textbook = 'Zumdahl'
number_of_terms = 30

import pandas as pd
from bs4 import BeautifulSoup
from googlesearch import search 
import requests
import re

# to search 
query = 'quizlet'+' '+textbook+' '+'chapter'+' '+chapter

links = []
for j in search(query, tld="co.in", num=40, stop=40, pause=3): 
    links.append(j)

headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:80.0) Gecko/20100101 Firefox/80.0'}

sub_dict = {'question': [], 'answer': []}

for url in links:
    soup = BeautifulSoup(requests.get(url, headers=headers).content, 'html.parser')


    for i, (question, answer) in enumerate(zip(soup.select('a.SetPageTerm-wordText'), soup.select('a.SetPageTerm-definitionText')), 1):
        sub_dict['question'].append(question.get_text(strip=True, separator='\n'))
        sub_dict['answer'].append(answer.get_text(strip=True, separator='\n'))

df = pd.DataFrame(sub_dict)
df = df.apply(lambda x: x.astype(str).str.lower())
df_final = df.groupby(['question'])['answer'].count().reset_index(name='count')
df_ff = df.merge(df_final, on = 'question', how = 'inner').sort_values(['count'], ascending=False)
df_ff = df_ff.drop_duplicates(subset='question', keep='first', inplace=False)
df_num = df_ff.head(number_of_terms)
df_qs = df_ff[df_ff['question'].str.contains("\?")]
df_qs = df_qs.head(number_of_terms)
```
